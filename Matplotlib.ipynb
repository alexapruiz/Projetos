**Importando as bibliotecas**

```
import matplotlib.pyplot as plt
import numpy as np
```

```
%matplotlib inline
```

**Definindo os valores de x e y da função**

```
x = np.linspace(0, 5 , 11)

y = x * x
```

```
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Gráfico da função x')
```

```
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Gráfico da função x')
plt.plot(x,y)
```

```
plt.plot(x,y,'r--')
```

```
plt.subplot(1, 2, 1)
plt.plot(x,y,'r--')

plt.subplot(1, 2, 2)
plt.plot(x,y,'g*')
```

```
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.3,0.3])
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
axes1.plot(x,y)
axes2.plot(y,x)
```

```
x = np.linspace(0, 5 , 11)
y = x * x

fig, axes = plt.subplots(figsize=(7,4), dpi=140)
axes.plot(x , y , 'r--', label = 'Dados1')
axes.plot(y , x , 'g*', label = 'Dados2')

axes.set_title('Gráfico')
axes.legend(loc=0)
```
